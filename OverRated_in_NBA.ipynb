# 2 - Modelagem dos Dados

## 2.1 - Seleção das Features

```python
colunas_excluir = ['year',
                   'name',
                   'playerId',
                   'playerSlug',
                   'teamId',
                   'status',
                   'avgAssists',
                   'avgBlocks',
                   'avgFieldGoalsAttempted',
                   'avgFieldGoalsMade',
                   'avgFouls',
                   'avgFreeThrowsAttempted',
                   'avgFreeThrowsMade',
                   'avgMinutes',
                   'avgPoints',
                   'avgRebounds',
                   'avgSteals',
                   'avgThreePointFieldGoalsAttempted',
                   'avgThreePointFieldGoalsMade',
                   'avgTurnovers',
                   'position',
                   'birthdate',
                   'college',
                   'draftinfo',
                   'birthplace',
                   'team',
                   'htwt',
                   'experience',
                   'rankingSalary'
]
```

```python
dataset_nba = df2.drop(columns=colunas_excluir,axis=1)
dataset_nba.info()
```

## 2.2 - Amostragem Aleatoria

```python
from sklearn.model_selection import train_test_split

test_size = 0.3
random_state = 21

train_set, test_set = train_test_split(dataset_nba,
                                       test_size = test_size,
                                       random_state = random_state)
```

```python
train_set.info()
```

```python
test_set.info()
```

## 2.3 - Preparação dos Dados

```python
dataset_nba_features = train_set.drop(columns=['salary'],axis=1)
dataset_nba_target = train_set['salary'].copy()
```

```python
dataset_nba_features.info()
```

```python
dataset_nba_target
```

## 2.4 - Tratamento das Variaveis Categoricas

```python
dataset_nba_features_cat = dataset_nba_features[dataset_nba_features.columns[(dataset_nba_features.dtypes == 'object')]]
dataset_nba_features_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

colunas = dataset_nba_features.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

one_hot = make_column_transformer((
    OneHotEncoder(drop = 'if_binary'),[0]),
                                  remainder='passthrough',
                                  sparse_threshold=0)
```

```python
nba_features_transform = one_hot.fit_transform(dataset_nba_features)
```

```python
one_hot.get_feature_names_out(colunas)
```

```python
nba_features_transform = pd.DataFrame(nba_features_transform, columns=one_hot.get_feature_names_out(colunas))
nba_features_transform
```

# 3 - Seleção do Modelo de Aprendizado de Maquina

## 3.1 - Regressão Linear

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(nba_features_transform, dataset_nba_target)

print(lin_reg.get_params())
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

salary_predict = lin_reg.predict(nba_features_transform)
lin_reg_rmse = np.sqrt(mean_squared_error(dataset_nba_target, salary_predict))

print("RMSE: ",lin_reg_rmse)
```

## 3.2 - Árvore de Decisão

```python
from sklearn.tree import DecisionTreeRegressor

dec_tree_reg = DecisionTreeRegressor(max_depth =3, random_state=5)

dec_tree_reg.fit(nba_features_transform, dataset_nba_target)
print(dec_tree_reg.get_params())
```

```python
from sklearn.metrics import mean_squared_error

salary_predict = dec_tree_reg.predict(nba_features_transform)
dec_tree_rmse = np.sqrt(mean_squared_error(dataset_nba_target, salary_predict))

print("RMSE: ",dec_tree_rmse)
```

## 3.3 - Floresta Aleatória

```python
from sklearn.ensemble import RandomForestRegressor

rand_forest_reg = RandomForestRegressor()

rand_forest_reg.fit(nba_features_transform, dataset_nba_target)

print(rand_forest_reg.get_params())
```

```python
from sklearn.metrics import mean_squared_error

salary_predict = rand_forest_reg.predict(nba_features_transform)

rand_forest_reg_rmse = np.sqrt(mean_squared_error(dataset_nba_target, salary_predict))

print("RMSE: ",rand_forest_reg_rmse)
```

## 3.4 - Metricas do Modelo Escolhido (Floresta Aleatoria)

### 3.4.1 - R2
Representa o percentual da variância dos dados que é explicado pelo modelo.

```python
from sklearn.metrics import r2_score

salary_predict = rand_forest_reg.predict(nba_features_transform)

r2 = r2_score(dataset_nba_target, salary_predict)

print("R²:", r2)
```

### 3.4.2 - MSE - Erro Quadrático Médio
A soma da média de diferença entre o valor predito com o real.

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(dataset_nba_target, salary_predict)

print("MSE:", mse)
```

### 3.4.3 - MAE - Erro médio absoluto
A média da diferença absoluta entre os valores previstos pelo modelo e os valores observados.

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(dataset_nba_target, salary_predict)
print("MAE:", mae)
```

Diferença Entre MAE e MSE:
A maior diferença entre o MAE e o MSE (Erro Quadrático Médio) é que o MSE penaliza mais os erros maiores. O MSE é calculado elevando os erros ao quadrado e depois calculando a média.

### 3.4.4 - Coeficiente de Correlação de Pearson
O coeficiente mede a intensidade e a direção de relações lineares

```python
from scipy.stats import pearsonr

corr, _ = pearsonr(dataset_nba_target, salary_predict)

print("Coeficiente de Correlação de Pearson:", corr)
```

### 3.4.5 - Matriz Confusao

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

threshold = dataset_nba_target.mean()
predicted_classes = ["Low" if salary < threshold else "High" for salary in salary_predict]
true_classes = ["Low" if salary < threshold else "High" for salary in dataset_nba_target]

conf_matrix = confusion_matrix(true_classes, predicted_classes)

disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot()
plt.xlabel('Previsões')
plt.ylabel('Verdadeiros Valores')
plt.title('Matriz de Confusão')
plt.show()
```

```python
vn, fp, fn, vp = confusion_matrix(true_classes, predicted_classes).ravel()
```

```python
vn, fp, fn, vp
```

### 3.4.6 - Acuracia
Diz quanto o meu modelo acertou das previsões possíveis.

```python
acuracia = (vp+vn)/(vn+vp+fp+fn)
print('Acuracia: ',acuracia)
```

### 3.4.7 - Precisao
A proporção de identificações positivas que foi realmente correta.

```python
precisao = (vp)/(vp+fp)
print('Acuracia: ', precisao)
```

### 3.4.8 - Recall
Quão bom meu modelo é para prever positivos.

```python
recall = (vp)/(vp+fn)
print('Acuracia: ', recall)
```

### 3.4.9 - F1 Score
Mostra o balanço entre a precisão e o recall de nosso modelo

```python
# F1 Score
f1 = (2*precisao*recall)/(precisao+recall)
print('Acuracia: ', f1)
```

```python
import pickle

with open('modelo_onehotenc.pkl', 'wb') as arquivo:
  pickle.dump(one_hot, arquivo)

with open('modelo_random_forest.pkl', 'wb') as arquivo:
  pickle.dump(rand_forest_reg, arquivo)
```

```python
modelo_one_hot = pd.read_pickle('/content/modelo_onehotenc.pkl')
modelo_arvore = pd.read_pickle('/content/modelo_random_forest.pkl')
```

## 4.2 - Incluindo o Dado para Previsão do Salário

```python
df_dados = pd.read_csv('nba_stats.csv')
df_dados['salary'] = df_dados['salary'].str.replace('$','')
df_dados['salary'] = df_dados['salary'].str.replace(',','').astype(int)
posicao_filtro = ['Not Available','Guard-Forward']
df_dados = df_dados[~df_dados['position'].isin(posicao_filtro)]
```

## 4.3 - Aplicando os Dados no Modelo

```python
caracteristicas = df_dados.drop(colunas_excluir, axis=1)
caracteristicas = modelo_one_hot.transform(caracteristicas)
salario_previsto = modelo_arvore.predict(caracteristicas)
df_dados['salario_previsto'] = salario_previsto
df_dados['diferenca'] = df_dados['salary'] - df_dados['salario_previsto']
```

## 4.4 - Resultado

### 4.4.1 - Top 10 Mais Superfaturados

```python
df_dados.sort_values(by=['diferenca'], ascending=False).head(10)
```

### 4.4.2 - Scatter entre Salario Real e Previsto

```python
x = df_dados['salary'].values
y = df_dados['salario_previsto'].values

coeficientes = np.polyfit(x, y, 1)
linha_regressao = np.poly1d(coeficientes)
plt.scatter(x, y, edgecolor='black')
plt.plot(x, linha_regressao(x), color='red')
plt.xlabel('Salario Real')
plt.ylabel('Salario Previsto')
plt.title('Salario Real x Previsto')
plt.show()
```
